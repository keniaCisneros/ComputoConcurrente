```
import threading
from threading import Thread, RLock
from random import randint
import time

class Mutex:
    def __init__(self):
        self.esta_bloqueado = False
        self.RLock = RLock()
    
    def bloquear(self):
        self.RLock.acquire()
        
    def desbloquear(self):
        self.RLock.release()

class Tortuga(Thread):
    def __init__(self, mutex):
        self.pos = 1
        self.tiempo = 0
        self.mutex = mutex
        super(Tortuga, self).__init__()
        
    def avanzaRapido(self):
        self.pos += 3
        print(f'T posicion {self.pos}')
        
    def resvala(self):
        self.pos -= 6
        if self.pos < 0:
            self.pos = 1
            print('T posicion 1')
        else:
            print(f'T posicion {self.pos}')
        
    def avanzaLento(self):
        self.pos += 1
        print(f'T posicion {self.pos}')
        
    def run(self):
        global meta
        while True:
            x = randint(1,100)
            if x <= 50:
                self.avanzaRapido()
                #time.sleep(1)
            if 50 < x <= 70:
                self.resvala()
                #time.sleep(1)
            else:
                self.avanzaLento()
                #time.sleep(1)
            if meta == True:
                break
            if 70 <= self.pos:
                self.tiempo = time.time()
                self.mutex.bloquear()
                meta = True
                self.mutex.desbloquear()
                break

class Liebre(Thread):
    def __init__(self, mutex):
        self.pos = 1
        self.mutex = mutex
        self.tiempo = 0
        super(Liebre, self).__init__()
        
    def duerme(self):
        print(f'L posicion {self.pos}')
        
    def granSalto(self):
        self.pos += 9
        print(f'L posicion {self.pos}')
        
    def resvalonGrande(self):
        self.pos -= 12
        if self.pos < 1:
            self.pos = 1
            print('L posicion 1')
        else:
            print(f'L posicion {self.pos}')
        
    def pequenioSalto(self):
        self.pos += 1
        print(f'L posicion {self.pos}')
        
    def resvalonPequenio(self):
        self.pos -= 2
        if self.pos < 1:
            self.pos = 1
            print('L posicion 1')
        else:
            print(f'L posicion {self.pos}')
        
    def run(self):
        global meta
        while True:
            x = randint(1,100)
            if x <= 20:
                self.duerme()
                #time.sleep(1)
            if 20 < x <= 40:
                self.granSalto()
                #time.sleep(1)
            if 40 < x <= 50:
                self.resvalonGrande()
                #time.sleep(1)
            if 50 < x <= 80:
                self.pequenioSalto()
                #time.sleep(1)
            else:
                self.resvalonPequenio()
                #time.sleep(1)
            if meta == True:
                break
            if 70 <= self.pos:
                self.tiempo = time.time()
                self.mutex.bloquear()
                meta = True
                self.mutex.desbloquear()
                break

meta = False
mutex = Mutex()

tortuga = Tortuga(mutex)
liebre = Liebre(mutex)
    
print('Inicio carrera')
    
tortuga.start()
liebre.start()

tortuga.join()
liebre.join()

if tortuga.tiempo < liebre.tiempo:
  print('Gano T')
elif liebre.tiempo < tortuga.tiempo:
  print('Gano L')
elif tortuga.tiempo == liebre.tiempo:
  print('Empate')        
```

```
import numpy as np
import time

x = 100
m1 = np.random.random((x, x))
m2 = np.random.random((x, x))

c1 = m1.dot(m2)
print(c1)
```

## Secuencial

```
t1 = time.time()
c2 = np.zeros((x, x))
for i in range (x):
  for j in range (x):
    for k in range (x):
      c2[i][j] += m1[i][k] * m2[k][j]
t2 = time.time()
print(c2)
print(t2 - t1)
```

## Hilos

```
import threading
from threading import Thread, RLock
import time
from queue import Queue
class Mutex:
    def __init__(self):
        self.esta_bloqueado = False
        self.RLock = RLock()
    
    def bloquear(self):
        self.RLock.acquire()
        
    def desbloquear(self):
        self.RLock.release()

def mult(mutex):
  global cola
  global m1
  global m2
  global f
  while cola.not_empty():
    mutex.bloquear()
    l = cola.get()
    mutex.desbloquear()
    for i in range (x):
      for j in range (x):
        f[l][j] += m1[l][i] * m2[j][l]

x = 100
m1 = np.random.random((x, x))
m2 = np.random.random((x, x))
cola = Queue(x)
f = np.zeros((x,x))
mutex = Mutex()

for i in range(x):
  cola.put(i)

th1 = time.time()

h1 = threading.Thread(target = mult, args = (mutex,))
h2 = threading.Thread(target = mult, args = (mutex,))
h3 = threading.Thread(target = mult, args = (mutex,))
h4 = threading.Thread(target = mult, args = (mutex,))

h1.start()
h2.start()
h3.start()
h4.start()

h1.join()
h2.join()
h3.join()
h4.join()

th2 = time.time()
print(f)
print(th2 - th1)
```
